# Creating a Time Series Plot using Pandas & Matplotlib
**Much, much more on this here: http://pandas.pydata.org/pandas-docs/stable/timeseries.html#timeseries **

```python
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

** Load some data that is published in a monthly format (Average Precipitation Output)**

```python
ao = np.loadtxt(r'../data/monthly.ao.index.b50.current.ascii')
```

**Load some dates to use as the index**

```python
dates = pd.date_range('1950-01', '2016-07', freq='M')
```

```python
dates
```

### Shapes Must Match! 

```python
dates.shape
```

```python
ao.shape
```

```python
average_output = pd.Series(ao[:,2], index=dates)
```

** And Now we can Plot **

```python
average_output.plot()
```

**Filter by key in the Series**

```python
average_output['2000':'2015'].plot()
```
